# Christmas Tree in matplotlib

## Leaves & Decorations

```python
# Preparing stuffs

import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# render leaves

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection="3d")

k = 600

Z = [(i/k)**1.5 + 0.4 for i in range(k)]
X = [math.cos(i / 4) * (k - i) / 1.75 for i in range(k)]
Y = [math.sin(i / 4) * (k - i) / 1.75 for i in range(k)]
ax.scatter(X, Y, Z, c="green", marker="^")


plt.xlim(-500, 500)
plt.ylim(-500, 500)
plt.show()
```

```python
# render decorations

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection="3d")

k = 300

step = 4
Z = [(i/k)**1.5 + 0.4 for i in range(1, k, step)]
X = [math.cos(i / 5 + 5) * (k - i + 10) for i in range(1, k, step)]
Y = [math.sin(i / 5 + 5) * (k - i + 10) for i in range(1, k, step)]
ax.scatter(X, Y, Z, c="red", marker="o")
ax.scatter([0], [0], [1.5], c="gold", marker="*", s=500)

plt.xlim(-500, 500)
plt.ylim(-500, 500)
plt.show()
```

```python
# render altogether

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection="3d")

k = 600

Z = [(i/k)**1.5 + 0.4 for i in range(k)]
X = [math.cos(i / 4) * (k - i) / 1.75 for i in range(k)]
Y = [math.sin(i / 4) * (k - i) / 1.75 for i in range(k)]
ax.scatter(X, Y, Z, c="green", marker="^")

k = 300

step = 4
Z = [(i/k)**1.5 + 0.4 for i in range(1, k, step)]
X = [math.cos(i / 5 + 5) * (k - i + 10) for i in range(1, k, step)]
Y = [math.sin(i / 5 + 5) * (k - i + 10) for i in range(1, k, step)]
ax.scatter(X, Y, Z, c="red", marker="o")
ax.scatter([0], [0], [1.5], c="gold", marker="*", edgecolor='black', s=500)

plt.xlim(-500, 500)
plt.ylim(-500, 500)
plt.show()
```

## Animation

```python
from matplotlib import rc, animation
rc('animation', html='html5')

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection="3d")

def render(f):
    fig.clear()
    ax = fig.add_subplot(111, projection="3d")
    
    k = 600

    Z = [(i/k)**1.5 + 0.4 for i in range(k)]
    X = [math.cos(i / 4 + (f / 10)) * (k - i) / 1.75 for i in range(k)]
    Y = [math.sin(i / 4 + (f / 10)) * (k - i) / 1.75 for i in range(k)]
    ax.scatter(X, Y, Z, c="green", marker="^")

    k = 300

    step = 4
    Z = [(i/k)**1.5 + 0.4 for i in range(1, k, step)]
    X = [math.cos(i / 5 + 5 + (f / 10)) * (k - i + 10) for i in range(1, k, step)]
    Y = [math.sin(i / 5 + 5 + (f / 10)) * (k - i + 10) for i in range(1, k, step)]
    ax.scatter(X, Y, Z, c="red", marker="o")
    ax.scatter([0], [0], [1.5], c="gold", marker="*", edgecolor='black', s=500)

    plt.xlim(-500, 500)
    plt.ylim(-500, 500)
    return fig

def init():
    return render(0),

def animate(f):
    return render(f),

ani = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=50, blit=True)
ani.save("christmas_tree.mp4")
ani
```
